```python
import pandas as pd
import random

#Walmart
Product_ID = [f"WA{str(i).zfill(3)}" for i in range(1, len(Product_Names) + 1)]
Min_Price = [round(random.uniform(0.5, 7.0), 2) for _ in range(len(Product_Names))]
Max_Price = [round(min_price + random.uniform(0.5, 3.0), 2) for min_price in Min_Price]

data1 = {
    'Product_Code': Product_ID,
    'Product_Category': Product_Category,
    'Product_Name': Product_Names,
    'Min_Price': Min_Price,
    'Max_Price': Max_Price,
    'Store': ['Walmart']*len(Product_Names)
}

df_walmart = pd.DataFrame(data1)

#Target
Product_ID = [f"TA{str(i).zfill(3)}" for i in range(1, len(Product_Names) + 1)]
Min_Price = [round(random.uniform(0.5, 9.0), 2) for _ in range(len(Product_Names))]
Max_Price = [round(min_price + random.uniform(0.5, 3.0), 2) for min_price in Min_Price]

data2 = {
    'Product_Code': Product_ID,
    'Product_Category': Product_Category,
    'Product_Name': Product_Names,
    'Min_Price': Min_Price,
    'Max_Price': Max_Price,
    'Store': ['Target']*len(Product_Names)
}

df_target = pd.DataFrame(data2)

#Aldi
Product_ID = [f"AL{str(i).zfill(3)}" for i in range(1, len(Product_Names) + 1)]
Min_Price = [round(random.uniform(0.5, 8.0), 2) for _ in range(len(Product_Names))]
Max_Price = [round(min_price + random.uniform(0.5, 3.0), 2) for min_price in Min_Price]

data3 = {
    'Product_Code': Product_ID,
    'Product_Category': Product_Category,
    'Product_Name': Product_Names,
    'Min_Price': Min_Price,
    'Max_Price': Max_Price,
    'Store': ['Aldi']*len(Product_Names)
}

df_aldi = pd.DataFrame(data3)
```

```python
import random

# Function to calculate average with random variation
def average_with_variation(value1, value2, variation_percent):
    average = (value1 + value2) / 2
    variation = average * (variation_percent / 100)
    random_offset = random.uniform(-variation, variation)
    result = average + random_offset
    return result

variation_percent = 10
current_price=[]

for i,j in zip(result_df['Min_Price'],result_df['Max_Price']):
    average=average_with_variation(i,j,variation_percent)
    current_price.append(round(average,2))

result_df['Current_Price']=current_price
```

```python
import random

def apply_random_discount(value, max_discount_percent):
    discount_percent = random.uniform(0, max_discount_percent)
    discounted_value = value - (value * discount_percent / 100)
    return discounted_value

discounts=[]
for price,store in zip(master_df['Current_Price'],master_df['Store']):
    if store=='Walmart':
        max_discount_percent=8
    elif store=='Target':
        max_discount_percent=3
    else:
        max_discount_percent=5
    discounted_price=apply_random_discount(price,max_discount_percent)
    discounts.append(round(discounted_price,2))

master_df['Discount_Price']=discounts
```

```python
import requests
from googleapiclient.discovery import build

def search_google_images(api_key, cx, query):
    service = build("customsearch", "v1", developerKey=api_key)
    res = service.cse().list(q=query, cx=cx, searchType='image').execute()

    if 'items' in res:
        first_result = res['items'][0]
        image_url = first_result['link']
        return image_url
    else:
        return None

def download_image(image_url, save_path):
    try:
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
        response = requests.get(image_url, headers=headers, stream=True)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=128):
                    f.write(chunk)
            return True
        else:
            print(f"Failed to download the image. Status code: {response.status_code}")
            return False
    except Exception as e:
        print(f"Error during image download: {str(e)}")
        return False


if __name__ == "__main__":
    # Replace 'YOUR_API_KEY' and 'YOUR_CX' with your API key and custom search engine ID
    api_key = 'AIzaSyAH9Lmtq6LEX-ob1tugkot-yq8cIV0zPmA'
    cx = 'c4e086388e09d4532'
    query = 'milk'

    image_url = search_google_images(api_key, cx, query)

    if image_url:
        print(f"The first image result for '{query}' is: {image_url}")

        # Download and save the image
        save_path = f"images/{query.replace(' ', '_')}.jpg"
        success = download_image(image_url, save_path)

        if success:
            print(f"Image downloaded and saved at: {save_path}")
        else:
            print("Failed to download the image.")
    else:
        print(f"No image found for '{query}'")
```
